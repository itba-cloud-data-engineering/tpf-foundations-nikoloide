```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine

csv_url = 'https://infra.datos.gob.ar/catalog/sspm/dataset/175/distribution/175.1/download/tipos-de-cambio-historicos.csv'
data = pd.read_csv(csv_url)
engine = create_engine('postgresql://username:secret@localhost:5432/database')
with engine.connect() as conn:
    print(bool(conn)) # <- just to keep track of the process
    conn.execute("TRUNCATE TABLE P_SALARIES.IPC_MES;")
    data.to_sql(name='P_SALARIES.IPC_MES', con=engine, if_exists='append')
    print("end") # <- just to keep track of the process
```

```python
engine = create_engine('postgresql://username:secret@localhost:5432/database')

url_dict = {'https://infra.datos.gob.ar/catalog/sspm/dataset/175/distribution/175.1/download/tipos-de-cambio-historicos.csv':'usd_mes',
           'https://cdn.produccion.gob.ar/cdn-cep/datos-por-actividad/salarios/w-median/w_median_privado_mensual_por_clae6.csv': 'ft_salarios_actividad',
           'https://cdn.produccion.gob.ar/cdn-cep/clae_agg.csv': 'claes',
           'https://infra.datos.gob.ar/catalog/sspm/dataset/145/distribution/145.3/download/indice-precios-al-consumidor-nivel-general-base-diciembre-2016-mensual.csv': 'ipc_mes'}

if __name__ == '__main__':
    print('load DB...')
    print('Connection to DB...')
    conn = engine.connect()
    print('Truncate tables...')
    conn.execute("TRUNCATE TABLE p_salaries.lk_fecha, p_salaries.ipc_mes, p_salaries.ft_salarios_actividad, p_salaries.claes;")
    print('Truncate tables OK')
    csv_dict = {'olist_customers_dataset': 'order_customer', 'olist_orders_dataset': 'orders', 'olist_order_payments_dataset': 'order_payments', 'olist_products_dataset': 'products', 'olist_order_items_dataset': 'order_items'}    
    for csv_url in url_dict:
            data = pd.read_csv(csv_url)
            with engine.connect() as conn:
                print(bool(conn)) # <- just to keep track of the process
                data.to_sql(name='P_SALARIES.' + url_dict[csv_url] , con=engine, if_exists='append', index=False)
                print("end") # <- just to keep track of the process
    conn.close()
    print('Closing connection to DB...')
```
